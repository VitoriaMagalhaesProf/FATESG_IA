```python
#Aluna: Vitória da Silva de Magalhães
#Banco: https://www.kaggle.com/datasets/algozee/teenager-menthal-healy?resource=download

import pandas as pd
# Importação de dados CSV com separador de colunas pelo caracter ","
dataset = pd.read_csv('Teen_Mental_Health_Dataset.csv', sep=',')
dataset
```

```python
# ** Qual é a contagem total dos acessos agrupados por sexo
dataset.groupby('gender')['gender'].count()
```

```python
# ** Qual é a idade máxima dos acessos
dataset['age'].max()
```

```python
# **Qual é a rede social mais usada?
dataset['platform_usage'].value_counts()
```

```python
#**Niveis de interação social baixo;
dataset[dataset['social_interaction_level'] == 'low']
```

```python
# **Idade mais atingida por baixa interação social, por mal uso das redes sociais.
dataset['age'].value_counts()
```

```python
# Histograma com as idades dos acessos realizados
dataset['age'].hist(bins=15, color='purple') #bins=50 (grossura das colunas)
```

```python
#Gráfico de Pizza que mostra o total por sexo
dataset['gender'].value_counts().plot(kind='pie', title='gender', legend=True, table=True, autopct='%1.1f%%')
```
