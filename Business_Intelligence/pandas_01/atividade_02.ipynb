```python
#Nivel 1 - Criar serie vazia;
import pandas as pd

serie_vazia = pd.Series()

print(serie_vazia.dtype)
```

```python
#Nivel 2 - Criar serie a partir de lista
import pandas as pd

frutas = ['graviola', 'banana', 'uva', 'pêssego', 'romã']

serie_frutas = pd.Series(frutas)

print(serie_frutas)
```

```python
#Nivel 3 - Índices personalizados
import pandas as pd
import numpy as np

valores = np.array([10, 20, 30, 40])

serie = pd.Series(valores, index=['A', 'B', 'C', 'D'])

print(serie)
```

```python
#Nivel 4 - Dicionario e NaN
import pandas as pd

dados = {'jan': 100, 'fev': 200}

serie = pd.Series(dados, index=['jan', 'fev', 'mar'])

print(serie)
```
